```python
import numpy as np
import pandas as pd
```

```python
cli, sup = [pd.read_csv(i) for i in ('train_clinical_data.csv', 'supplemental_clinical_data.csv') ]
```

```python
for df in [cli, sup]:
    df['visit_month_diff'] = df.groupby(['patient_id'])['visit_month'].diff()
    df['visit_month_diff_min'] = df.groupby('patient_id')['visit_month_diff'].transform('min')
```
